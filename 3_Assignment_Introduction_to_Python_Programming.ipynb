## Assignment 1: Basic If-Else Statement

**Task:** Write a Python program that asks the user to input their age and prints whether they are eligible to vote (18 years or older) or not.

```python
p = int(input("Enter the age"))
if p >= 18:
    print("Eligible")
else:
    print("Not Eligible")
```

## Assignment 2: Multiple Conditions

**Task:** Write a Python program that asks for the user's age and prints the appropriate message:
- If the user is 18 or older, print 'Adult'
- If the user is 13 or older but under 18, print 'Teenager'
- If the user is younger than 13, print 'Child'

```python
p = int(input("Enter the age"))
if(p >= 18):
    print("Adult");
elif(p >= 13):
    print("Teenager");
else:
    print("Child");
```

## Assignment 3: Nested If-Else

**Task:** Write a Python program that checks if a user’s input number is positive, negative, or zero.
- If the number is positive, print 'Positive Number'
- If the number is negative, print 'Negative Number'
- If the number is zero, print 'Zero'

```python
p = int(input("Enter the Number"))
if(p > 0):
    print("Positive Number");
elif(p < 0):
    print("Negative Number");
else:
    print("Zero");
```
